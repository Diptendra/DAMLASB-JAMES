## 1.1 Enter a string and create a table with all the characters in a string in alphabetical orders

```python
def getKey(item):
    return item[0]
input_string=raw_input('Enter a line : \n')
input_string=input_string.lower().replace(" ","")
char_counts = dict()
for char in input_string:
    if char not in char_counts:
        char_counts[char] = 1
    else:
        char_counts[char] = char_counts[char] + 1
final_table = list()
for key, val in char_counts.items():
    final_table.append( (key,val) )
final_table=sorted(final_table, key=getKey)
for key, val in final_table[:]:
    print key, val
```

## HW1.4 Longest word in Alice in wonderland

```python
import sys, re, string
def getKey(item):
    return item[0]
file_name = raw_input('Enter the file name: ')
try:
    file_hand = open(file_name)
except:
    print 'File cannot be opened:', fname
    exit()
word_counts = dict()
for line in file_hand:
    regex = re.compile('[%s]' % re.escape(string.punctuation))
    #replace each punctuation character with a space
    line = regex.sub(' ', line.lower())  
    #replace multiple spacees with a single space
    line = re.sub( '\s+', ' ', line)
    line = re.sub("\d+", "", line)
    words = line.split()
    for word in words:
        
        if word not in word_counts:
            word_counts[word] = len(word)
final_lst = list()
for key, val in word_counts.items():
    final_lst.append( (val,key) )
final_lst=sorted(final_lst, key=getKey, reverse=True)
print(final_lst[0])            
    
```
